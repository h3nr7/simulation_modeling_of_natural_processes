# General introduction to dynamic systems

1. System that varies through time
2. System is described by its state $ s(t) $
3. state $s$ can be multivariate (a vector)

    $ s(t) = (s_1(t),...s_n(t))^T $
    

#### Decrete dynamical system are described by recurrence relations

$ s(t + \Delta t) = f(s(t)) $

#### Continuous dynamical system are described by *differential equations*

$ \dot{s} \equiv \large\frac{ds}{dt} \small = f(s(t)) $


## Generality of the description

#### Continuous case

What happens if we have second derivatives

$ \dot{s} + \ddot{s} = f(s) $

Let us then define

$ \dot{s} = y $

$ \dot{y} = f(s) - y $


By defining $u(t) \equiv (s(t), y(t)) $, we have a system of the form

$ \dot{u} = g(u) $


#### Discrete case

"Second derivatives" equivalent to presence of a $ t - \Delta t $ term

$ s(t + \Delta t) = f(s(t)) + s(t - \Delta t) $

Similar too continuous case

$ y(t + \Delta t) = s(t) $

$ s(t + \Delta t) = f(s(t)) + y(t) $

and by defining $ u(t) \equiv (s(t), y(t)) $ we have a system of the form

$ u(t + \Delta t) = g(u(t)) $


#### Continuous case where $ f $ depends on $s$ as well as $t$

$ \dot{s} = f(s, t) $

Let us then define

$ \dot{s} = f(s, y) $

$ \dot{y} = 1 $

with $ y(0) = 0 $, by defining $ u(t) \equiv (s(t), y(t)) $ we have a system of the form

$ \dot{u} = g(u) $
